# Congruencia Lineal

### Numeros primos relativos entre si

```python
import pandas as pd
import math

n = 30
df = pd.DataFrame(index = range(n+1), columns = ['Xn','Un'])
a = 12121212
b = 5
m = int(math.pow(2,48) - 1)
x0 = 100000

df.iloc[0] = [x0,'*']

for i in range(n): 
    xn = (a*x0 + b) % m; 
    ui = xn/m
    df.iloc[i+1] = [xn,ui]
    x0 = xn
df
```
